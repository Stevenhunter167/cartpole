```python
import numpy as np
import matplotlib.pyplot as plt
import gym

from IPython.display import clear_output
import time

from learner import Learner
from util import *

np.random.seed(0)
```

```python
env = gym.make('CartPole-v0')
learner = Learner()
BESTREWARD = 50
```

```python
# train using initial data
data = np.loadtxt("g50.txt")
learner.train(DATA=data, epochs=300, verbose=2)
```

```python
# policy iteration 10 epochs
def policyIteration(epochs, exploration, batch_size=10):
    avg_rwds = []
    for i in range(epochs):
        # evaluate 10 games
        rwd, avg_act = evaluate(env, learner, 5, render=True)
        # report current agent performance
        avg_rwd = np.mean(rwd)
        avg_rwds.append(avg_rwd)
        print(f"iteration {i} avg_rwd {avg_rwd}")
        # create training data
        data, acc_rwd, rwd = generateDataBatch(env, learner, size=batch_size, exploration=exploration, criteria=avg_rwd)
        # train the policy network
        learner.train(DATA=data, epochs=3, verbose=2)
        clear_output(wait=True)
    return avg_rwds
```

```python
policyIteration(5, exploration=0.15)
```

```python
# avgrwd on each iteration 160, 166.6, 186.6, 192.4, 198.8
learner.model.save('198_8.model')
```

```python
# cartpole is solved after 6 iterations
policyIteration(10, exploration=0.15)
```
